```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('World population data.csv')
```

```python
data = data.dropna()
```

## Handling missing values

- Our dataset originally had **273 rows**.
- After dropping rows with null values, we got **266 rows**.  
- That means only **7 rows (≈ 2.5%)** had missing values.  

**Dropping rows is fine** if we lose only a small fraction of data (e.g., 2.5%).  
❌ If many rows had nulls, it’s usually better to **fill missing values** instead of dropping.

```python
# select  first 10 rows
subset = data[['Country Name', '2022 [YR2022]']].head(10)
```

```python
subset
```

```python

#Bar chart
plt.figure(figsize=(10,5))
plt.bar(subset['Country Name'], subset['2022 [YR2022]'], color='skyblue')
plt.title('Population Distribution by country in 2022')
plt.xlabel('Country')
plt.xticks(rotation=90)
plt.ylabel('Population')
plt.show()
```

```python
Top_10_country = data[['Country Name', '2022 [YR2022]']].sort_values(ascending=False,by='2022 [YR2022]').head(10)
Top_10_country
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=Top_10_country['2022 [YR2022]'], y=Top_10_country['Country Name'], data = Top_10_country)
plt.title("Top 10 Countries by Population in 2022")
plt.xlabel("Population")
plt.ylabel("Country")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.histplot(Top_10_country, x='2022 [YR2022]', bins=True, kde=True)
```
